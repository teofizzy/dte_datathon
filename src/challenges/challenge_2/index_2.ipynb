# Overview



## Clinical research
Clinical research is medical research that examines humans to better comprehend health and illness. Clinical research helps enhance the way doctors treat and prevent illness. Through clinical research, scientists discover:

* How the organism functions
* How disease progresses in humans, including how diseases improve or deteriorate over time
* How the body reacts to a potential treatment
* Which behaviors promote health and illness prevention, and which behaviors increase the likelihood of becoming ill?
* The objective is to use science to enhance the health care and health of individuals over time. Clinical research studies may or may not provide participants with personal benefits.

### Clinical trial
Clinical trials are research studies where participants are randomly assigned to experimental interventions, often without physicians' prescription. These trials are also known as interventional investigations, allowing participants to evaluate their experiences without choosing the intervention.

### Participants of a clinical research
Researchers seek individuals who meet a certain set of eligibility criteria. These criteria specify who can and cannot participate in a study and may include:

Individuals of a particular age or gender
Those with or without a particular malady, disease, or health condition.
Individuals with or without a particular health history, such as a prior treatment, will be evaluated.
People who are subjected to or are in contact with something that affects their health
Researchers use eligibility criteria to ensure the safety of participants and enroll the appropriate individuals in order to collect the data necessary to answer the research question. There are numerous varieties of research studies, each with distinct eligibility requirements.

Feel free to read more about clinical research [here](https://clinicaltrials.gov/study-basics/learn-about-studies)

## Data Description
`NCT Number`: A unique identification code given to each clinical study record registered on ClinicalTrials.gov. The format is "NCT" followed by an 8-digit number (for example, NCT00000419). Also called the ClinicalTrials.gov identifier.

`Study Title`: Title for the different clinical studies.

`Study URL`: Link to the website containing all the details of the study.

`Acronym`: The acronym or initials used to identify a clinical study (not all studies have one).

`Study Status`: Indicates the current recruitment status or the expanded access status.

`Study Results`: The progress of the study whether it's completed and submitted to ClinicalTrials.gov (yes) or it's incomplete so no results posted (no).

`Conditions`: The disease, disorder, syndrome, illness, or injury that is being studied. On ClinicalTrials.gov, conditions may also include other health-related issues, such as lifespan, quality of life, and health risks.

`Interventions`: A process or action that is the focus of a clinical study. Interventions include drugs, medical devices, procedures, vaccines, and other products that are either investigational or already available. Interventions can also include noninvasive approaches, such as education or modifying diet and exercise.

`Primary Outcome Measures`: In a clinical study's protocol, the planned outcome measure that is the most important for evaluating the effect of an intervention/treatment. Most clinical studies have one primary outcome measure, but some have more than one.

`Secondary Outcome Measures`: In a clinical study's protocol, a planned outcome measure that is not as important as the primary outcome measure for evaluating the effect of an intervention but is still of interest. Most clinical studies have more than one secondary outcome measure.

`Sponsors`: The organization or person who initiates the study and who has authority and control over the study.

`Collaborators`: An organization other than the sponsor that provides support for a clinical study. This support may include activities related to funding, design, implementation, data analysis, or reporting.

`Sex`: A type of eligibility criteria that indicates the sex of people who may participate in a clinical study (all, female, male). Sex is a person's classification as female or male based on biological distinctions. Sex is distinct from gender-based eligibility.

`Age`: A type of eligibility criteria that indicates the age a person must be to participate in a clinical study. This may be indicated by a specific age or the following age groups:
- Child (birth-17)
- Adult (18-64)
- Older Adult (65+)

`Phases`: The stage of a clinical trial studying a drug or biological product, based on definitions developed by the U.S. Food and Drug Administration (FDA). The phase is based on the study's objective, the number of participants, and other characteristics. There are five phases: Early Phase 1 (formerly listed as Phase 0), Phase 1, Phase 2, Phase 3, and Phase 4. Not Applicable is used to describe trials without FDA-defined phases, including trials of devices or behavioral interventions.

`Enrollment`: The number of participants in a clinical study. The "estimated" enrollment is the target number of participants that the researchers need for the study.

`Funder Type`: Describes the organization that provides funding or support for a clinical study. This support may include activities related to funding, design, implementation, data analysis, or reporting. Organizations listed as sponsors and collaborators for a study are considered the funders of the study. ClinicalTrials.gov refers to four types of funders:
- U.S. National Institutes of Health
- Other U.S. Federal agencies (for example, Food and Drug Administration, Centers for Disease Control and Prevention, or U.S. Department of Veterans Affairs)
- Industry (for example: pharmaceutical and device companies)
- All others (including individuals, universities, and community-based organizations)

`Study Type`: Describes the nature of a clinical study. Study types include interventional studies (also called clinical trials), observational studies (including patient registries), and expanded access.

`Study Design`: The investigative methods and strategies used in the clinical study.
- Allocation: A method used to assign participants to an arm of a clinical study. The types of allocation are randomized allocation and nonrandomized.
  - Randomized allocation: A type of allocation strategy in which participants are assigned to the arms of a clinical trial by chance.
- Intervention model:The general design of the strategy for assigning interventions to participants in a clinical study. Types of intervention models include: single group assignment, parallel assignment, cross-over assignment, and factorial assignment.
  -Single group assignment: A type of intervention model describing a clinical trial in which all participants receive the same intervention/treatment.
  -Parallel assignment: A type of intervention model describing a clinical trial in which two or more groups of participants receive different interventions. For example, a two-arm parallel assignment involves two groups of participants. One group receives drug A, and the other group receives drug B. So during the trial, participants in one group receive drug A "in parallel" to participants in the other group, who receive drug B.
  -Cross-over assignment:A type of intervention model describing a clinical trial in which groups of participants receive two or more interventions in a specific order. For example, two-by-two cross-over assignment involves two groups of participants. One group receives drug A during the initial phase of the trial, followed by drug B during a later phase. The other group receives drug B during the initial phase, followed by drug A. So during the trial, participants "cross over" to the other drug. All participants receive drug A and drug B at some point during the trial but in a different order, depending on the group to which they are assigned.
  -Factorial Assignment: A type of intervention model describing a clinical trial in which groups of participants receive one of several combinations of interventions. For example, two-by-two factorial assignment involves four groups of participants. Each group receives one of the following pairs of interventions: (1) drug A and drug B, (2) drug A and a placebo, (3) a placebo and drug B, or (4) a placebo and a placebo. So during the trial, all possible combinations of the two drugs (A and B) and the placebos are given to different groups of participants.
- Masking:A clinical trial design strategy in which one or more parties involved in the trial, such as the investigator or participants, do not know which participants have been assigned which interventions. Types of masking include: open label, single blind masking, and double-blind masking.
  - Investigator: A researcher involved in a clinical study.
- Primary purpose: The main reason for the clinical trial. The types of primary purpose are: treatment, prevention, diagnostic, supportive care, screening, health services research, basic science, and other.


`Primary Completion Date`: The date on which the last participant in a clinical study was examined or received an intervention to collect final data for the primary outcome measure. Whether the clinical study ended according to the protocol or was terminated does not affect this date. For clinical studies with more than one primary outcome measure with different completion dates, this term refers to the date on which data collection is completed for all the primary outcome measures. The "estimated" primary completion date is the date that the researchers think will be the primary completion date for the study.

`Completion Date`: The date on which the last participant in a clinical study was examined or received an intervention/treatment to collect final data for the primary outcome measures, secondary outcome measures, and adverse events (that is, the last participant's last visit). The "estimated" study completion date is the date that the researchers think will be the study completion date.

`First Posted`: The date on which the study record was first available on ClinicalTrials.gov after National Library of Medicine (NLM) quality control (QC) review has concluded. There is typically a delay of a few days between the date the study sponsor or investigator submitted the study record and the first posted date.

`Results First Posted`: The date on which summary results information was first available on ClinicalTrials.gov after National Library of Medicine (NLM) quality control (QC) review has concluded. There is typically a delay between the date the study sponsor or investigator first submits summary results information (the results first submitted date) and the results first posted date. Some results information may be available at an earlier date if Results First Posted with QC Comments.

`Last Update Posted`: The most recent date on which changes to a study record were made available on ClinicalTrials.gov. There may be a delay between when the changes were submitted to ClinicalTrials.gov by the study's sponsor or investigator (the last update submitted date) and the last update posted date.

`Locations`: Information on where this study is being conducted.

`Study Documents`: Refers to the type of documents that the study sponsor or principal investigator may add to their study record. These include a study protocol, statistical analysis plan, and informed consent form.

`Other IDS`: Identifiers or ID numbers other than the NCT number that are assigned to a clinical study by the study's sponsor, funders, or others. These numbers may include unique identifiers from other trial registries and National Institutes of Health grant numbers.

`Start date`: The actual date on which the first participant was enrolled in a clinical study. The "estimated" study start date is the date that the researchers think will be the study start date.

### Countries
The study has taken note of the fact that country names come immediately before the pipe operator '|' and at the end of the string. As such the study will design a program to extract the country names where each clinical trial was conducted

```python
# Function to get country names from text
def get_country_names(text):
    # Initialize geonamescache
    gc = geonamescache.GeonamesCache()

    # Get a set of country names
    country_names = set(country_data['name'] for country_data in gc.get_countries().values())

    # Split the text by '|'
    parts = text.split('|')

    # Initialize an empty list to store country names
    extracted_countries = set()

    for part in parts:
        # Check if a part contains a comma
        if ',' in part:
            # Split the part by comma and get the last part
            country = part.split(',')[-1].strip() # Remove leading and trailing whitespace
            if country in country_names:
                extracted_countries.add(country)

    return list(extracted_countries)
```

```python
# Apply the function to the DataFrame
heart_df["countries"] = heart_df["locations"].apply(get_country_names)

# Display the DataFrame with extracted countries
heart_df.head()
```

### Text Analytics

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Define a text preprocessing function
def preprocess_text(text):
    # Expand contractions - words like didn't to did not
    expanded_text = contractions.fix(text)

    # Remove words with less than 3 letters
    words = expanded_text.split()
    filtered_words = [word for word in words if len(word) >= 3]

    # Remove noise
    cleaned_words = [re.sub(r'[^A-Za-z0-9-]+', '', word) for word in filtered_words]

    # Convert digit characters to words
    final_words = [str(w2n.word_to_num(word) if word.isdigit() else word) for word in cleaned_words]

    # Join the words back into a cleaned text
    cleaned_text = ' '.join(final_words)

    # Tokenize, lemmatize, and remove stopwords
    doc = nlp(cleaned_text)

    lemmatized_tokens = [token.lemma_.strip().lower() for token in doc if not token.is_stop and len(token.text) > 2]

    # Join the words back into a cleaned text
    cleaned_text = ' '.join(lemmatized_tokens)

    return cleaned_text
```

```python
# Get the corpus
corpus = heart_df['brief_summary'].apply(preprocess_text)
```

```python
# Define a function to get ngrams
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx])
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10] # Return the top 10

# The most common words
top_unigrams = get_top_ngram(corpus, 1)

# Visualize the top 10 unigrams
x1, y1=map(list,zip(*top_unigrams))
sns.barplot(x=y1,y=x1, palette='viridis')
```

```python
# top 10 bigrams
top_bigrams = get_top_ngram(corpus, 2)

# Visualize the top 10 bigrams
x2, y2=map(list,zip(*top_bigrams))
sns.barplot(x=y2,y=x2, palette='viridis')
```

```python
# top 10 trigrams
top_trigrams = get_top_ngram(corpus, 3)
x3, y3=map(list,zip(*top_trigrams))
sns.barplot(x=y3,y=x3, palette='viridis')
```

```python
# wordcloud
def plot_word_cloud(corpus):
    """This function creates a word cloud for the text corpus"""
    text = ' '.join(corpus)  # Combine the corpus into a single string
    wordcloud = WordCloud(
        max_words=400,
        width=3000,
        height=1500,
        background_color='black',
        stopwords=STOPWORDS).generate(text)
    fig = plt.figure(
        figsize=(14, 8),
        facecolor='k',
        edgecolor='k')
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()

# Use the function with your corpus
word_cloud = plot_word_cloud(corpus)

# show word cloud
word_cloud
```

## Bivariate Data Analysis
> In this phase, the study shall inspect the relationship between two variables at a time.

```python
# Finding out the association between categorical columns
## filter categorical columns
categorical_data= heart_df.select_dtypes(include='category')

# Cramer's V function
def cramers_v(var1, var2):
    crosstab = pd.crosstab(var1, var2)
    chi2 = chi2_contingency(crosstab)[0]
    n = var1.count()
    phi2 = chi2 / n
    r, k = crosstab.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))
col_list = (list(categorical_data.columns))[1:]
rows = []
for var1 in col_list:
    col = []
    for var2 in col_list:
        cramers = cramers_v(categorical_data[var1], categorical_data[var2])
        col.append(round(cramers, 2))
    rows.append(col)

cramers_results = np.array(rows)
cramers_df = pd.DataFrame(cramers_results, columns=col_list, index=col_list)

# Set the size of the heatmap
plt.figure(figsize=(15, 10))

# Create a heatmap using Seaborn
sns.heatmap(cramers_df, annot=True, cmap=sns.color_palette("ch:start=.2,rot=-.3"), linewidths=0.5)

# Set the title of the heatmap
plt.title("Association of the Categorical Variables")

# Display the heatmap
plt.show()
```

#### How to interpret:
Cramér’s V is an effect size measurement for the chi-square test of independence. It measures how strongly two categorical fields are associated.

- ES ≤ 0.2	The result is weak. Although the result is statistically significant, the fields are only weakly associated.

- 0.2 < ES ≤ 0.6	The result is moderate. The fields are moderately associated.

- ES > 0.6	The result is strong. The fields are strongly associated.

>> For example there is a strong association of 0.75 between the funder type and the sponsors of a study.

```python
# Function to plot a bar chart for two columns
def plot_grouped_bar_chart(df, x_column, y_column, top_n=10, hue=None):
    grouped_data = df.groupby(x_column)[y_column].mean().nlargest(top_n)
    sns.barplot(x=grouped_data.values, y=grouped_data.index, palette='viridis', hue=hue)
    plt.title(f'{x_column} vs {y_column}')
    plt.xlabel(y_column)
    plt.ylabel(x_column)
    plt.show()

# Function to plot relationships between two numerical columns
def plot_relationship(df, x_column, y_column):
    sns.scatterplot(x=x_column, y=y_column, data=df, palette='viridis')
    plt.title(f'{x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
```

```python
# sponsors that participated in at least 50 studies
def plus_50_studies(df, n=50):
    # Count the number of studies for each sponsor
    sponsor_counts = df['sponsor'].value_counts().reset_index()
    sponsor_counts.columns = ['sponsor', 'study_count']

    # Filter for sponsors with at least n studies
    qualified_sponsors = sponsor_counts[sponsor_counts['study_count'] >= n]['sponsor']

    # Filter the original DataFrame to include only qualified sponsors
    selected_data = df[df['sponsor'].isin(qualified_sponsors)]

    return selected_data

# create a function to filter the dataframe
def filter_category(df, cat_col, num_col):
  return df.groupby(cat_col)[num_col].mean().reset_index()

# Finding the top n
def find_top(df, num_col, n):
  sorted_df = df.sort_values(by=num_col, ascending=False)
  sorted_df = sorted_df.head(n)
  return sorted_df

def plot_top_categories(df, category_col, num_col, n):
    """
    Plot the top N categories based on a numerical column
    """
    top_categories = df.nlargest(n, num_col)
    sns.set_palette("viridis")

    plt.figure(figsize=(12, 6))
    plt.barh(top_categories[category_col], top_categories[num_col])
    plt.title(f'Top {n} {category_col} by {num_col}')
    plt.xlabel(f'{num_col}')
    plt.ylabel(f'{category_col}')
    plt.gca().invert_yaxis()  # Invert the y-axis to show the top categories at the top
    plt.show()

# using the functions

## sponsors that have participated in at least 50 studies
selected_sponsors = plus_50_studies(heart_df)

## Group by sponsor for mean number of countries
sponsors_countries = filter_category(selected_sponsors, 'sponsor', 'num_countries')
```
